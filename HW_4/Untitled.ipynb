```python
import time, random, math
import numpy as np
import gym
```

```python
env = gym.make('Taxi-v2')
env.reset()
env.render()

RECORD = None
MAX_EPISODES = 1000001
MAX_STEPS = env.spec.timestep_limit
EPSILON = 1
EPS_DECR = 0.0000005
DISCOUNT = 0.9
LEARNING_RATE = 0.03
```

```python
n_states = env.observation_space.n
n_actions = env.action_space.n
q_table = np.zeros((n_states, n_actions))
n_visits = np.zeros((n_states, n_actions))

T = np.zeros((n_states, n_actions, n_states))
R = np.zeros((n_states, n_actions))

def updateQTable(prevState, prevAction, reward, curState, done):
    future = DISCOUNT*max(q_table[curState]) if not done else 0
    q_table[prevState][prevAction] += LEARNING_RATE*(reward + future - q_table[prevState][prevAction])


def getAction(curState):
    if random.random() < EPSILON:
        a = env.action_space.sample()
    else:
        a = np.argmax(q_table[curState])
    n_visits[curState][a] += 1
    return a
    
```

```python

# obsMin = 0
# obsMax = env.observation_space.n
# actionMin = 0
# actionMax = env.action_space.n

totalreward = 0

for episode in range(MAX_EPISODES):
    
    if episode%1000 == 0 :
        print("Episode =", episode, " |  EPS =", EPSILON, " |  Avg Reward =", totalreward/1000, " | Q[462,4] =", q_table[462,4])
        totalreward = 0

    EPSILON -= EPS_DECR
    curState = env.reset()
    for step in range(MAX_STEPS):
        #env.render()
        prevState = curState
        action = getAction(curState)
        curState, reward, done, info = env.step(action)
        totalreward += reward
        updateQTable(prevState, action, reward, curState, done)
        if done :
            break
```
